## Objective

The objective of this project is to analyze and visualize the Iris dataset using different graphical techniques. The project aims to understand the distribution of features, identify relationships between variables, and compare the different Iris species through visual representations such as histograms, scatter plots, and bar charts.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['SepalLengthCm'].hist()
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'])

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')

plt.show()
```

```python
species_count = df['Species'].value_counts()

species_count.plot(kind='bar')

plt.title('Count of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')

plt.show()
```

## Conclusion

The Iris dataset was analyzed using various visualization techniques. A histogram was used to understand the distribution of sepal length, a scatter plot was used to identify relationships between sepal length and petal length, and a bar chart was used to compare species counts. The analysis revealed a positive relationship between sepal length and petal length, and the dataset was found to be balanced across all three species.
